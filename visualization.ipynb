```python
import matplotlib.pyplot as plt
import os
import numpy as np
from utils.data_loader import JsonReader
from PIL import Image
import json
```

```python
images_list = os.listdir('./data/images/')
```

```python
def load_json(filename):
    with open(filename, 'r') as f:
        data = json.load(f)
    return data

def show_image(image_path):
    image = Image.open(os.path.join('./data/images/', image_path))
    plt.imshow(np.asarray(image))
```

```python
data = load_json('./results/train.json')
```

```python
for each in images_list:
    if each == '.DS_Store':
        continue
    show_image(each)
    plt.show()
    print(each)
    print("Generation:{}".format(data[each]['GT']))
    print('Ground Truth:{}'.format(data[each]['Pred']))
```
